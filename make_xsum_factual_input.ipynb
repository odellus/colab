```python
summaries = ds_train['summary']
support_docs = ds_train['support']
labels = ds_train['is_factual']
input_texts = [
    f'{doc} {tokenizer.sep_token} {summary}' for (doc, summary) 
    in zip(support_docs, summaries)
]
ds_input = datasets.Dataset.from_dict({'input_text': input_texts, 'label': labels})
```
